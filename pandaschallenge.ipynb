```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "02-Homework_04-Pandas_Instructions_PyCitySchools_Resources_schools_complete.csv"
student_data_to_load = "02-Homework_04-Pandas_Instructions_PyCitySchools_Resources_students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Make a Dataframe of the total number of schools, students, total budget
# We will also include the Avg. Math & Reading scores
# Finally, we'll calculate the % of students with a passing math grade, passing reading grade, and then overall passing %

    # School Count

school_count = len(school_data["School ID"].unique())    
    
    #Student Count

student_count = len(student_data["Student ID"].unique())

    #Total Budget
school_data['Total Budget'] = school_data.iloc[:,4:].sum(axis=1)

total_budget = sum(school_data['Total Budget'])

    #Average Math Score
    
student_data['Average Math Score'] = student_data.iloc[:,5:].mean(axis=1)

average_math = student_data['Average Math Score'].mean()

    #Average Reading Score

student_data['Average Reading Score'] = student_data.iloc[:,4:].mean(axis=1)

average_reading = student_data['Average Reading Score'].mean()

    # % of Students with Passing Math Grade

total_math_passing2 = student_data.loc[student_data["math_score"]>70]

total_math_passing1=total_math_passing2.iloc[:,6:].count()

total_math_passing = (total_math_passing1 / student_count)*100


    # % of Students with Passing Reading Grade
    
total_reading_pass2 = student_data.loc[student_data["reading_score"]>70]

total_reading_pass1=total_reading_pass2.iloc[:,5:].count()

total_reading_pass = (total_reading_pass1 / student_count)*100


    # Run DataFrame

    
first_Dataframe = pd.DataFrame(
    
                               {"Total Schools" :school_count,
                                "Student Count" : student_count,
                                "Total Budget": total_budget,
                                "Average Math Score": average_math,
                                "Average Reading Score": average_reading,
                                "Percent Passed - Math": total_math_passing,
                                "Percent Passed - Reading": total_reading_pass})


first_Dataframe.head(1)


```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Establish Variables & Build Dataframe

    # Determine School Types

school_n_types = school_data.set_index(["school_name"])["type"]

school_n_types

    # School Size 

school_size = school_data.set_index(["school_name"])["size"]

school_size

    # Total Budget

total_budget_school = school_data.set_index(["school_name"])["budget"]

total_budget_school


    # Spending Per Student

spending_per_student = total_budget_school / school_size

spending_per_student
    

    # Avg. Math & Reading Score per School

math_per_school = school_data_complete.groupby(["school_name"]).mean()["math_score"]
reading_per_school = school_data_complete.groupby(["school_name"]).mean()["reading_score"]


    # % Passing Math & Reading per School
        
        # Total Amount of Passing Students 
        
math_passing_per = school_data_complete[(school_data_complete["math_score"]>70)]
reading_passing_per = school_data_complete[(school_data_complete["reading_score"]>70)]

        # Total Students per School

total_pass_math_per_school = math_passing_per.groupby(["school_name"]).count()["student_name"]

total_pass_reading_per_school = reading_passing_per.groupby(["school_name"]).count()["student_name"]

        # Percent of Students

per_pass_math_school = (total_pass_math_per_school / school_size) * 100

per_pass_reading_school = (total_pass_reading_per_school / school_size) * 100

        # Percent that Passed Both

per_pass_both = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]

        # Group By School

per_pass_both = per_pass_both.groupby(["school_name"]).count()["student_name"]

percent_pass_both_school = (per_pass_both / school_size) * 100

        # DataFrame It
    
dataframe_school_sum = pd.DataFrame({
    
                                    "School Type": school_n_types,
                                    "School Size": school_size,
                                    "Total Budget": total_budget_school,
                                    "Spending Per Student": spending_per_student,
                                    "Average Math Score": math_per_school,
                                    "Average Reading Score": reading_per_school,
                                    "% Passing Math": total_pass_math_per_school,
                                    "% Passing Reading": total_pass_reading_per_school,
                                    "% Passing Both": percent_pass_both_school})

dataframe_school_sum
    
```
